```python
import matplotlib.pyplot as plt  
import numpy as np  
import math as math
import scipy
```

```python
def comb(a,b):  
    return math.factorial(a)/(math.factorial(b)*math.factorial(a-b))
```

```python
comb(10, 4)
```

```python
n = 10
p = 0.5
x = np.arange(0, n, 1)  
y = [comb(n, k) * (p**k) * ((1-p) ** (n-k)) for k in x]
plt.plot(x,y)  
plt.xlabel('x')  
plt.ylabel('y')  
plt.title('Lab DLS') 
slice = n // 3 + 1
plt.fill_between(x[5:5+slice], y[5:5+slice], color = "blue")
plt.show()
```
