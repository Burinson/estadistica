```python
# Jesús Uriel Guzmán Mendoza

# Ejercicio 1

from matplotlib import pyplot as plt
import numpy as np  
import math as math
import random

def comb(a, b):  
    return math.factorial(a) / (math.factorial(b) * math.factorial(a - b))

def binomial(x, n, p):
    b = [comb(n, k) * (p**k) * ((1 - p)**(n - k)) for k in x]
    return b

def graficaSuave(x, y, titulo, show = True):
    plt.plot(x, y)
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show() if show else 0

def graficaBarras(x, y, titulo, show = True):
    plt.bar(x, y)
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show() if show else 0

def graficaPuntos(x, y, titulo, show = True):
    plt.scatter(x, y)
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show() if show else 0

def graficaLineas(x, y, titulo, show = True):
    z = range(0, len(y), 5)
    plt.plot(z, y[::5])
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show() if show else 0
```

```python
# Ejercicio 1
n = 100
m = 5
p = [random.randint(1, 99) / 100.0 for i in range(m)] # m probabilidaes a lo we, m x 4 gráficas en totaly
p.sort()

x = range(0, n, 1)
print(p)

for i in range(len(p)):
    y = binomial(x, n, p[i])

    # a. Usando una gráfica suave (como la hecha en clase).
    graficaSuave(x, y, 'Gráfica suave de distribución binomial({},{})'.format(n, p[i]))

    # b. Usando una gráfica de barras.
    graficaBarras(x, y, 'Gráfica de barras de distribución binomial({},{})'.format(n, p[i])) 

    # c. Usando una gráfica de puntos.
    graficaPuntos(x, y, 'Gráfica de puntos de distribución binomial({},{})'.format(n, p[i]))

    # d. Usando una gráfica de líneas.
    graficaLineas(x, y, 'Gráfica de lineas de distribución binomial({},{})'.format(n, p[i]))
```

```python
#Pregunta 3, inciso a
n = 1000
p = 0.60
x = range(0, n, 1)
y = binomial(x, n, p)

#Pregunta 3, inciso b y c
secciones = [(0, 12, "red"), (0, 300, "blue"), (360, 850, "green"), (111, 999, "yellow")]

graficaSuave(x, y, 'Gráfica suave de distribución binomial({},{})'.format(n, p), False)
for seccion in secciones:
    f = 0
    for punto in y[seccion[0]:seccion[1]]:
        f += punto
    plt.fill_between(x[seccion[0]:seccion[1]],y[seccion[0]:seccion[1]], color = seccion[2])    
    print('{} < x < {} = {}'.format(seccion[0], seccion[1], f))
plt.show()


graficaBarras(x, y, 'Gráfica de barras de distribución binomial({},{})'.format(n, p), False)
for seccion in secciones:
    f = 0
    for punto in y[seccion[0]:seccion[1]]:
        f += punto
    plt.fill_between(x[seccion[0]:seccion[1]],y[seccion[0]:seccion[1]], color = seccion[2])    
    print('{} < x < {} = {}'.format(seccion[0], seccion[1], f))          
plt.show()

graficaPuntos(x, y, 'Gráfica de puntos de distribución binomial({},{})'.format(n, p), False)
for seccion in secciones:
    f = 0
    for punto in y[seccion[0]:seccion[1]]:
        f += punto
    plt.fill_between(x[seccion[0]:seccion[1]],y[seccion[0]:seccion[1]], color = seccion[2])    
    print('{} < x < {} = {}'.format(seccion[0], seccion[1], f))
plt.show()

graficaLineas(x, y, 'Gráfica de lineas de distribución binomial({},{})'.format(n, p), False)
for seccion in secciones:
    f = 0
    for punto in y[seccion[0]:seccion[1]]:
        f += punto
    plt.fill_between(x[seccion[0]:seccion[1]],y[seccion[0]:seccion[1]], color = seccion[2])    
    print('{} < x < {} = {}'.format(seccion[0], seccion[1], f))  
plt.show()
```
