```python
from matplotlib import pyplot as plt
import numpy as np  
import math as math
import random

def comb(a, b):  
    return math.factorial(a) / (math.factorial(b) * math.factorial(a - b))

def binomial(x, n, p):
    b = [comb(n, k) * (p**k) * ((1 - p)**(n - k)) for k in x]
    return b

def graficaSuave(x, y, titulo):
    plt.plot(x, y)
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

def graficaBarras(x, y, titulo):
    plt.bar(x, y)
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

def graficaPuntos(x, y, titulo):
    plt.scatter(x, y)
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

def graficaLineas(x, y, titulo):
    z = range(0, len(y), 5)
    plt.plot(z, y[::5])
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
n = 100
m = 3
p = [random.randint(1, 99) / 100.0 for i in range(m)] # m probabilidaes a lo we, m x 4 gráficas en totaly
p.sort()

x = range(0, n, 1)
print(p)

for i in range(len(p)):
    y = binomial(x, n, p[i])
    graficaSuave(x, y, 'Gráfica suave de distribución binomial({},{})'.format(n, p[i]))
    graficaBarras(x, y, 'Gráfica de barras de distribución binomial({},{})'.format(n, p[i]))
    graficaPuntos(x, y, 'Gráfica de puntos de distribución binomial({},{})'.format(n, p[i]))
    graficaLineas(x, y, 'Gráfica de lineas de distribución binomial({},{})'.format(n, p[i]))
```

```python
y = binomial(x, n, 0.43)
slice = n // 3 +1

for i in range(1, 4):
    plt.plot(x, y)
    plt.fill_between(x[(i - 1) * slice: i * slice], y[(i - 1) * slice: i * slice], color = "blue")
    plt.title('Gráfica suave de distribución binomial({},{})'.format(n, 0.43))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

# Para la la región que cubre el primer tercio de la gráfica:
    # Esta región representa la probabilidad de un evento cuando tomamos en consideración
    # todas las x ∈ [0, 34] para n = 100 y p = 0.43. Se puede observar que la región
    # pintada tiene una probabilidad muy baja de ocurrir, en donde la máxima probabilidad
    # ocurre aproximadamente en 34 y P(34) en b(34; 100; 0.43) ≈ 0.01 o 1%

# Para la la región que cubre el segundo tercio de la gráfica:
    # Esta región representa la probabilidad de un evento cuando tomamos en consideración
    # todas las x ∈ (34, 67] para n = 100 y p = 0.43. Se puede observar que la región
    # pintada tiene la probabilidad más alta de ocurrir, ya que esta región cubre todo
    # lo que viene siendo el centro de la distribución binomial. La máxima probabilidad
    # en esta región occurre aproximadamente en x ≈ 43, en donde P(42) ≈ 0.08 o 8%

# Para la la región que cubre el tercer tercio de la gráfica:
    # Esta región representa la probabilidad de un evento cuando tomamos en consideración
    # todas las x ∈ (67, 100) para n = 100 y p = 0.43. Se puede observar que la región 
    # pintada viene siendo la región con menor probabilidad (o nula probabilidad) de
    # toda la distribución, en donde cualquier valor de P(x) para cualquier x ∈ (67, 100) 
    # viene siendo ≈ 0%

```
