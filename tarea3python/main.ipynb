```python
from IPython.display import HTML, display
import matplotlib.pyplot as plt

def displayTable(data, columnNames):
    html = "<br><table style=\"border:1px solid white\">"
    html += "<thead>"
    for name in columnNames:
        html += "<td style=\"border:1px solid white; padding: 10px; text-align: center\"><h3>%s</h3></td>"%(name)
    html += "</thead> <body>"
    for row in data:
        html += "<tr>"
        for field in row:
            html += "<td style=\"border:1px solid white; padding: 10px; text-align: center\">%.2f</td>"%(field)
        html += "</tr>"
    html += "</body></table><br>"
    display(HTML(html))

def calc(x, y, fix = 2):
    data = []
    labels = ["x", "y", "x - Mx", "y - My", "(x - Mx)^2", "(x - Mx)(y - My)"]

    n = len(x)
    Mx = 0
    My = 0
    Sx2 = 0
    Sxy = 0

    for i in range(n):
        Mx = Mx + x[i] / n # promedio de x
        My = My + y[i] / n # promedio de y

    for i in range(n):
        data.append([]) # nueva fila

        data[i].append(x[i])
        data[i].append(y[i])
        data[i].append(x[i] - Mx)
        data[i].append(y[i] - My)
        data[i].append((x[i] - Mx)**2)
        data[i].append((x[i] - Mx) * (y[i] - My)) # se calcula cada uno de los valores

        Sx2 = Sx2 + (x[i] - Mx)**2 / n # Sx^2
        Sxy = Sxy + ((x[i] - Mx) * (y[i] - My)) / n # Sxy
    
    m = Sxy / Sx2 # pendiente de la recta
    b = My - m * Mx # intercepto de la recta
    
    
    displayTable(data, labels) # muestra la tabla

    print("Mx = %.2f" % Mx)
    print("My = %.2f" % My)
    print("Sx^2 = %.2f" % Sx2)
    print("Sxy = %.2f" % Sxy)
    print("y = %.2f x + %.2f" % (m, b))
    print("") # y alguna datos relevantes

    newy = []

    for i in range(n):
        newy.append(m * x[i] + b) # se calcula la nueva y para la regresión lineal

    labels = ["Datos originales", "Regresión lineal"] # títulos de las gráficas

    original, = plt.plot(x, y) # datos originales
    linReg, = plt.plot(x, newy) # datos de la regresión lineal

    handles = [original, linReg] 

    plt.legend(handles, labels)
    plt.show()

```

```python
x = [47, 62, 65, 70, 70, 78, 95, 100, 114, 118]
y = [38, 62, 53, 67, 84, 79, 93, 106, 117, 116]

calc(x, y)
```
